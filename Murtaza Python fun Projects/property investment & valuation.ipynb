```python
print("--- Property Investment Pro v1.0 ---")
# choice = input('''\n Options Select Kijea \n Buy press (1) \n Sale press (2) \n Main Menu press (0)\n''')
# Inputs ((Buy karne ki sorat main)Variables aur Type Casting )
buy_price = float(input("Ghar kitne mein liya tha? (e.g. 5500000): "))
renovation_cost = float(input("Mazeed kitna kharcha hua? (e.g. 1000000): "))
current_demand = float(input("Ab kitne ka bechna chahte hain? (e.g. 8000000): "))
monthly_rent = float(input("Abhi wahan se mahina rent kitna aa raha hai?: "))

# Calculations
total_investment = buy_price + renovation_cost
profit = current_demand - total_investment

# Rental Yield Calculation (Bohat zaroori step)
# Formula: (Salana Rent / Total Cost) * 100
annual_rent = monthly_rent * 12
rental_yield = (annual_rent / total_investment) * 100

# Lists ka use: Recommendations store karne ke liye
results = []
results.append(f"Total Investment: {total_investment}")
results.append(f"Expected Profit: {profit}")
results.append(f"Rental Yield: {round(rental_yield, 2)}%")

# Output
print("\n--- Final Analysis ---")
print(results[0]) # List indexing practice[cite: 3]
print(results[1])
print(results[2])


# print(result)
# Decision Logic (If-Else Practice[cite: 4])
if rental_yield >= 5:
    print("Mashwara: Yeh aik achi investment hai, rent acha mil raha hai.")
elif rental_yield >= 3:
    print("Mashwara: Investment theek hai, lekin market value barhne ka intezar karein.")
else:
    print("Mashwara: Rent bohot kam hai, bechna behtar ho sakta hai.")

# print(result)
```
