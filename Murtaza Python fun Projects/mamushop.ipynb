```python
mamushop = [{'item':'floor', 'price':100 , 'inventory': 250},
            {'item':'sugar', 'price':180 , 'inventory': 500},
            {'item':'rice', 'price':300 , 'inventory': 500},
            {'item':'oil', 'price':500 , 'inventory': 1000},
            {'item':'ghee', 'price':500 , 'inventory': 500},
            {'item':'tea', 'price':400 , 'inventory': 100},
            {'item':'soap', 'price':200 , 'inventory': 50},
            {'item':'salt', 'price':50 , 'inventory': 100}]
```

```python
item_name = input("kia Chahaye????")
for a in mamushop:
    if a['item'] == item_name:
        print(f"{item_name} {a['price']} ka hay")
        quantity = float(input("Kitna Chahiye hy?"))
        if a['inventory'] >= quantity:
            pass
        else:
            print(f"{item_name} itna nhi hy ")
            break
else:
    print(f"{item_name} nhi hy")
            
```
