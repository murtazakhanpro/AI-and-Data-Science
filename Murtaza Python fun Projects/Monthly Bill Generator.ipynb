```python
    print("Welcome To LYNTICA Management System")
    while True:
        # 1. Login System
        user_name = input('Please enter your username: ')
        Password = input('Please enter your password: ')
    
        if user_name == "admin" and Password == '123':
            print("[✔️] Access Granted!")
            # break
            # 2. Floor Selection & Rent Assignment
            # .lower() use kiya hai taake capital/small letters ka masla na ho
            while True:
                floor = input("Please Enter Your Floor ( g / 1 / 2 ): ").lower()
            
                if floor == 'g':
                    rent = 13500
                    security_deposit_status = input("Security Deposit (Y / N)")
                    if security_deposit_status == 'n':
                        sec_dep_sta = "Pending"
                        security_deposit = int(input("Please Enter Security Deposit Amount.\n"))
            
                    else:
                        sec_dep_sta = 'Paid'
                        security_deposit = int(input("Please Enter Security Deposit Amount.\n"))
            
                elif floor == '1':
                    rent = 14000
                    security_deposit_status = input("Security Deposit (Y / N)")
                    if security_deposit_status == 'n':
                        sec_dep_sta = "Pending"
                        security_deposit = int(input("Please Enter Security Deposit Amount.\n"))
            
                    else:
                        sec_dep_sta = 'Paid'
                        security_deposit = int(input("Please Enter Security Deposit Amount.\n"))
                elif floor == '2':
                    rent = 15000
                    security_deposit_status = input("Security Deposit (Y / N)")
                    if security_deposit_status == 'n':
                        sec_dep_sta = "Pending"
                        security_deposit = int(input("Please Enter Security Deposit Amount.\n"))
            
                    else:
                        sec_dep_sta = 'Paid'
                        security_deposit = int(input("Please Enter Security Deposit Amount.\n"))
                else:
                    print("[!] Apne Ghalat floor Darj Kia hai!")
                    rent = None 
            
                # 3. Tenant Details and Arrears Logic
                # Sirf tab chalega agar floor sahi ho[cite: 1]
                if rent is not None:
                    tenant_name = input("Please Enter Tenant Name: ")
                    contact = input("Kindly Enter Contact Number: ")
                    month = input("Kindly enter the month: ")
                    
                    # Bills and Payments
                    k_electric_input = input("Enter Electric Bill Amount (ya khali chor dein agar nahi aya): ")
            
                    if k_electric_input == "":
                        k_electric = 0.0
                        status_electric = "Bill Not Generated Yet!"
                    else:
                        k_electric = float(k_electric_input)
                        status_electric = "Generated"
                    
                    sui_gas_input = (input("Enter Sui Gas Bill Amount: \n Agar Bill Nhi Aya hai to khali Rehny Dijea "))
                    if sui_gas_input == "":
                        sui_gas = 0.00
                        status_gas = "Bill Not Generated Yet!"
                    else:
                        sui_gas = float(sui_gas_input)
                        status_gas = "Generated"
                    
                    
                    
                    
                    prev_balance = float(input("Enter any previous pending balance (or 0): "))
                    
                    # Total amount jo tenant ko pay karni thi
                    total_payable_amount = rent + k_electric + sui_gas + prev_balance + security_deposit
                    
                    print(f"\nTotal Amount Due: {total_payable_amount}")
                    cash_received = float(input(f"Enter cash received from {tenant_name}: "))
                    
                    # 4. Balance Calculation (Agay Wali Logic)
                    new_balance = total_payable_amount - cash_received
                    
                    if new_balance > 0:
                        status = f"PARTIAL PAID (Remaining: {new_balance})"
                    elif new_balance < 0:
                        status = f"ADVANCE PAID (Credit: {abs(new_balance)})"
                    else:
                        status = "FULL PAID"
            
                    # 5. Final Invoice Generation(jo mane ap copy past kr k tanant ko share karo gy)
                    # is line ko bhi thora phely print karwana hai kyun k jab hum whatsapp pr share kr rahe hain to 
                    #starting main bohat spacing aa rahi hai
                    print(f'''
                    ========================================================
                    LYNTICA BILL GENERATION SYSTEM
                    ========================================================
                    Tenant:           {tenant_name.upper()}
                    Contact:          {contact}
                    Month:            {month.capitalize()}
                    --------------------------------------------------------
                    {month.capitalize()} Rent:               {rent}
                    Electricity Bill:    {k_electric}        {status_electric}
                    Sui-Gas Bill:        {sui_gas}           {status_gas}
                    Previous Arrears:    {prev_balance}
                    Security Deposit:    {security_deposit}  {sec_dep_sta}
                    --------------------------------------------------------
                    TOTAL DUE:           {total_payable_amount}
                    CASH RECEIVED:       {cash_received}
                    --------------------------------------------------------
                    PAYMENT STATUS:      {status}
                    NEXT MONTH BALANCE:  {new_balance}
                    
                    ===== Thank You for Choosing Lyntica =====
                    ========================================================
                    ''')
                    choice = input("Kia Ap Aur Bill Generate Karna Chahaty hain").lower()
                    if choice != 'n':
                        print("Allah Hafiz")
                    break

                break

                    # Extra warning agar balance bacha ho
                    # if new_balance > 0:
                        # print(f"NOTE: Please remind {tenant_name} to clear {new_balance} PKR soon.")
        
        else:
            print("\n[❌] Access Denied: Wrong Credentials \n Please Try Again")
```
