### convensional style to declear variable in python
    - snake_case: Lowercase letters with underscores separating words. This is the standard for variables, functions, methods, and modules.
    
    -PascalCase (CapWords): Capitalized first letters for every word with no underscores. This is the standard for classes and exceptions.
    
    -SCREAMING_SNAKE_CASE: Uppercase letters with underscores separating words. This is used for constants defined at the module level.
