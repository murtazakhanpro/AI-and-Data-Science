# Conditions (IF / ELSE)

```python
age = int(input("Please enter your age: "))
if age >= 18:
    print("You Can Vote: ")
else:
    print("You Cannot Vote: ")
```

```python
print("Enter Your Marks")
marks = int(input("Enter your marks: "))

if marks >= 80:
    print("A Grade")
elif marks >= 60:
    print("B Grade")
elif marks >= 50:
    print("C Grade")
elif marks >= 40:
    print("D Grade")
else:
    print("Fail")
```

```python
num_1 = int(input("enter your number"))


if num_1 % 2 == 0:
    print("Even")
else:
    print("Odd")
```
