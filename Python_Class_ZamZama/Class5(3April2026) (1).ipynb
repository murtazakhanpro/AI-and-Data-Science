```python
english = float(input("Enter english marks: "))
computer = float(input("Enter computer marks: "))
mathematics = float(input("Enter mathematics marks: "))
urdu = float(input("Enter urdu marks: "))
sindi = float(input("Enter sindi marks: "))

total = english + computer + mathematics + urdu + sindi
percentage = total / 500 *100

if percentage>=90:
    grade = 'A+'
elif percentage>=80:
    grade = "A"
elif percentage>=70:
    grade = "B"
elif percentage>=60:
    grade = "C"
elif percentage>=50:
    grade = "D"
else:
    grade = "F"


print(f"""
         #######  Report #######
         
    English:       {english}
    Computer:      {computer}
    Mathematics:   {mathematics}
    Urdu:          {urdu}
    Sindi:         {sindi}
    
    Your tot marks {total} out of 500.
    
    Your perc is   {percentage}
    
    Your Grade is : {grade}
    
    
    """)
```

```python
mouse = int(input("The price of mouse is: "))
keyboard = int(input("The price of keyboard is: "))
headfone = int(input("The price of headfone is: "))
screen = int(input("The price of screen is: "))
router = int(input("The price of router is: "))

total= mouse+keyboard+headfone+screen+router
if total>50000:
    discount = total * 0.15
elif total>30000:
    discount = total * 0.10
elif total>20000:
    discount = total * 0.075
elif total>10000:
    discount = total * 0.05
else:
    discount = total * 0

print(f"""

         ######## Receipt ########
         
         Mouse       : {mouse}
         Keyboard    : {keyboard}
         Headfone    : {headfone}
         Screen      : {screen}
         Router      : {router}
         
         Total       : {total}
         Discount    : {discount}
         
         Payable Amt:  {total - discount}
""")
```
