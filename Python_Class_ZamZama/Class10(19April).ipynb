# Loops: 
   - For in   : it requires a collection to run over.
   - While    : it requres a condition to be true.

```python
for a in range(5):
    print(a,"nasir hussain")
```

# Continue and Break

```python
for a in range(10):
    if a==4:
        break
    else:
        print(a)
```

```python
for a in range(10):
    if a==4 or a==7 or a==0:
        continue
    else:
        print(a)
```

```python
for a in range(10):
    if a%2 ==0:
        print(f"{a} is even")
    else:
        print(f"{a} is odd")
```
