#### Take input of five subject marks.
    - eng
    - comp
    - math
    - urdu
    - sindi
    Calculate total marks
    Calclate percentage


```python
english = float(input("Enter english marks: "))
computer = float(input("Enter computer marks: "))
mathematics = float(input("Enter mathematics marks: "))
urdu = float(input("Enter urdu marks: "))
sindi = float(input("Enter sindi marks: "))

total = english + computer + mathematics + urdu + sindi
percentage = total / 500 *100

print(f"""
         #######  Report #######
         
    English:      {english}
    Computer:     {computer}
    Mathematics:  {mathematics}
    Urdu:         {urdu}
    Sindi:        {sindi}
    
    Your total marks are {total} out of 500.
    
    You percentage is {percentage}
    
    """)
```

```python
mobile = int(input("The price of mobile phone is: "))
laptop = int(input("The price of laptop is: "))
shoes = int(input("The price of shoes is: "))
jeans = int(input("The price of jeans is: "))
ipad = int(input("The price of ipad is: "))

total_amount = mobile + laptop+ shoes+jeans+ipad
tax  = total_amount*0.15

print(f"Your total amount is Rs: {total_amount}")
print(f"Your tax amount is Rs:{tax}")
print(f"Your payable amount is Rs: {total_amount + tax}")
```
