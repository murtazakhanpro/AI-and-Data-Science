# Slicing:
    - Copying a part of list.

```python
#index(+)      0        1       2       3      4          5        6      7 
students = ['Rehman', 'Umer', 'Ali', 'Hamza','Hassan', 'Yousuf','Ilyas','Hyder']
#index(-)    -8        -7      -6      -5      -4         -3      -2       -1
```

```python
students[2:6]
```

```python
students[3:7]
```

```python
students[:]
```

```python
students[:5]
```

```python
students[5:]
```

```python
students[-7:-2]
```

```python
students[0:8]
```

```python
students[0:8:1]  # start:end:step
```

```python
students[0:8:2]  # start:end:step
```

```python
students[0:8:4]  # start:end:step
```
