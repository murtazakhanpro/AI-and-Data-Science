```python
arr2d = np.arange(101,151).reshape(5,10)
arr2d
```

```python
arr2d > 130
```

```python

arr2d = np.random.random_integers(1,101, 100)
arr2d
```

```python
arr1d = np.random.random_integers(1,21,20)
arr1d
```

```python
arr1d [arr1d >= 10]
# is condition ma >= 10 select karega ga given array say
```

```python
arr1d [~(arr1d >= 10)]
# isme bhi condition same he rahe gi lakin, ~ (tilt) ka use krte 
# isko chor kr baqi sub select kar lega
```

```python
arr1d[arr1d%2==0]
# ishe array say sub even numbers select kardo
```

```python
arr1d[(arr1d>5) & (arr1d<10)]
# isme wo numbers ayin gy to 5 aur 10 k darmian ma hoon ge
```
