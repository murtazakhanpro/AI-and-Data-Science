# Goal Do Math

```python
num1 = int(input("Enter your First number: "))
num2 = int(input("Enter Your second number: "))

sum = num1 + num2
mul = num1 * num2
sub = num1 - num2
div = num1 / num2
print("Total is Addition:", sum)
print("Total is Subtraction:", sub)
print("Total is Multiplication:", mul)
print("Total is division:", div)
```

# Age ko input karwaya hai

```python
age = int(input("enter your age"))
days = age * 365
hours = days * 24
mint = hours * 60

print("Ap ab tak" , days , " din", hours , "ghanty aur itny" , mint , " minutes is duniya main reh chuky hain") 
```

#  Simple Calculator Banaya hai

```python
#print("simple mortage calculator")
loan = int(input("apne loan ki amount enter karain"))
total_time = int(input("ap kitne waqt main loan wapis karain gy years add karain"))
months = total_time * 12 
qist = loan / months

print("apka loan in amount hai", loan , "aur apki monthly installment hai", qist)
print("total loan" , qist * months)
```
