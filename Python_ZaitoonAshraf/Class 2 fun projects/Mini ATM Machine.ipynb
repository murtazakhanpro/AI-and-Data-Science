```python
print("Welcome To Murtaza Bank Of Pakistan:")

name = input("Enter Your Name: ")

pin = input("Enter Your PIN: ")
set_pin = input("Enter You New PIN: ")

if pin == set_pin:
    print("Access Denied")
else:
    balance = 52000

    amount = int(input("Enter withdraw amount: "))

    if amount > balance:
        print("Insufficient Funds")
    else:
        remaining_bal = balance - amount
        print(f"Hello {name}, your current balance is {remaining_bal}")

        desposit = input("For Deposit Press (y) For Exit Press(n): ")

        if desposit == "y":
            desposit_amount = int(input("Please Enter Your Amount: "))
            remaining_bal = remaining_bal + desposit_amount
            print(f"Hello {name}, your current balance is {remaining_bal}")
        else:
            print("Thanks For Using This ATM")
```
