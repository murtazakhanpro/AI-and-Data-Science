# order pizza

```python
availble_toppings = ['Mushroom', 'Olives', 'Onion', 'Pepricca', 'cheese', 'Capcicam']
applied_toppings = []
while True:
    topping = input("Please enter your toppin or q to exit")
    if topping != 'q':
        if topping in availble_toppings:
            applied_toppings.append(topping)
        else:
            print("Sorry! Requested topping is not availble")
    else:
        print("ThankYou For Ordering Pizza!")
        break
availble_toppings
```
