```python
# write a python function that takes a string as argument
# function analyze string: (len, alphabet, spaces, number, 
# lower, punctuation,)
# vowels, words, sentence, paragraph, 

# def uni_str(name)
```

```python
def tex_analyzer(txt):
    lenght = len(txt)
    # print(f"lenght of txt is{ lenght}")
    words = txt.split(" ")
    # print(f" Number of words in txt are{len(words)}")
    sentences = len(txt.split("."))
    # print(f" Number of words in txt are{len(sentences)}")
    paragraph = len(txt.split("\n"))
    # print(f"Number of sentence in txt are {len(paragraph)}")
    vow = 'aeiouAEIOU'  
    punc = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
    punctuation = 0
    alphabets = 0
    spaces = 0
    numbers = 0
    lower_case = 0
    upper_case = 0
    vowels=0
    others= 0
    for char in txt:
        if char.isalpha():
            alphabets+=1
            if char.upper():
                upper_case+=1
            else:
                lower_case+=1
            if char in vow:
                    vowels+=1
                    
            


        elif char.isspace():
            spaces+=1
        elif char.isnumeric():
            numbers+=1
       
        elif char in punc:
            punctuation+=1
        else:
            others+=1

    print(f'''
        Text Lenght :   {lenght}
        words  :        {words}
        sentenses are   {sentences}
        paragraph are:  {paragraph}
        upper case :    {upper_case}
        lower are :     {lower_case}
        spaces are :    {spaces}
        number are :    {numbers}
        alphabets are:  {alphabets}
        vowels :        {vowels}
        punctuations :  {punctuation}
        others   :      {others}
           ''')
```

```python
tex_analyzer('''
Dummy text is placeholder content used by designers, publishers, and developers to fill visual layouts before final content is ready.
It mimics the natural flow of real language so observers focus 
on fonts, alignment, and composition rather than 
reading the text. The standard format since the 1500s is Lorem Ipsum, a 
scrambled sequence of Latin words derived from Cicero's philosophical writings
             ''')
```
