# Return 

```python
def CalclateTax(salary=0, tax=0): # parameter
    if salary:
        if salary<600000:
            tax = salary*0  # local variable: scope within function
        elif salary<1200000:
            tax=salary*0.05
        elif salary<2400000:
            tax=salary*0.10
        elif salary<4800000:
            tax=salary*0.20
        else:        
            tax=salary*0.40
        
        return tax
    else:
        return tax *100
    
```

```python
calculated_tax = CalclateTax(1500000)
```

```python
print(calculated_tax)
```

```python
CalclateTax(tax=15000)
```

```python
CalclateTax(salary=10000000)
```

# Bank

```python
import random
bank_all_accounts = []
bank_all_closed_accounts = []
```

```python
def open_account(cnic, initial_deposit=0):
    account={'cnic':cnic,'balance':initial_deposit}
    account['account_title']= input('Enter account title: ')
    account['pin'] = random.randint(1000,9999)
    account['account_number'] = random.randint(1000000,9999999)
    bank_all_accounts.append(account)
    print('Thank you for opening account with us!')
    return account
```

```python
open_account(98765122,1000)
```

```python
bank_all_accounts
```

```python
def check_balance(account_number,pin):
    for acc in bank_all_accounts:
        if acc['account_number']==account_number:
            if acc['pin']==pin:
                return f"your current balance is Rs:{acc['balance']}"
            else:
                return "Invalid Pin"
    else:
        return f'{account_number }- Invalid Account'
```

```python
check_balance(6527477,8808)
```

```python
check_balance(9914629,5275)
```

```python
check_balance(9914629,5276)
```

```python
def cash_withdrawl(account_number, pin, amount):
    for acc in bank_all_accounts:
        if acc['account_number']==account_number:
            if acc['pin']==pin:
                if acc['balance']>=amount:
                    acc['balance']-=amount
                    return f"You withdrawl Rs:{amount} successfully"
                else:
                    return 'Insufficient Balance!'                
            else:
                return "Invalid Pin"    
    else:
        return f'{account_number }- Invalid Account'
```

```python
cash_withdrawl(6527477,8808,500)
```

```python
check_balance(6527477,8808)
```

```python
def cash_deposit(account_number, amount):
    for acc in bank_all_accounts:
        if acc['account_number']==account_number:
            acc['balance']+=amount
            return f"You Deposited Rs:{amount} successfully"
    else:
        return f'{account_number }- Invalid Account'
```

```python
cash_deposit(6527477,50000)
```

```python
check_balance(6527477,8808)
```

```python
def transfer(account_number, pin, amount, beneficiary_acc):
    for acc in bank_all_accounts:
        if acc['account_number']==account_number:
            if acc['pin']==pin:
                if acc['balance']>=amount:
                    for bacc in bank_all_accounts:
                        if bacc['account_number']==beneficiary_acc:
                            acc['balance']-=amount
                            bacc['balance']+=amount
                            return f"You transfered Rs:{amount} successfully!"
                    else:
                        return f'{account_number }- Invalid Beneficiary Account'
                else:
                    return 'Insufficient Balance!'                
            else:
                return "Invalid Pin"    
    else:
        return f'{account_number }- Invalid Account'
```

```python
transfer(6527477,8808, 50000, 4579508)
```

```python
check_balance(6527477,8808)
```

```python
def close_account(account_number, pin):
    for index, acc in enumerate(bank_all_accounts):
        if acc['account_number']==account_number:
            if acc['pin']==pin:                
                bank_all_closed_accounts.append(bank_all_accounts.pop(index))
                return f"Collect Rs:{acc['balance']}. You closed your account successfully!"                
            else:
                return "Invalid Pin"
    else:
        return "Account Not Found"
```

```python
close_account(6527477,8808)
```

```python
check_balance(6527477,8808)
```

```python
bank_all_closed_accounts
```

```python
def change_pin(account_number, pin, new_pin):
    for acc in bank_all_accounts:
        if acc['account_number']==account_number:
            if acc['pin']==pin:
                acc['pin'] = new_pin
                return f'Your pin changed Successfully!'
            
            else:
                return "Invalid Pin"
    else:
        return "Account Not Found"
```
