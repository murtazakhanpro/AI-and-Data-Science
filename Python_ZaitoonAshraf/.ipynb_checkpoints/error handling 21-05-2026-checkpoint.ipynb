```python
num1 = int(input("Please enter a number"))
num2 = int(input("Please enter a number"))
try:
    print(num1/num2)
except ZeroDivisionError:
    print("You can not divide a number by 0")
```

```python
lst = [1,2,3,4,5]
try:
    num1 = int(input("Please enter a number"))
    num2 = int(input("Please enter a number"))
    res = num1/num2
    print(res , lst[10])
except Exception as e:
    print(f"Exception caught : {e}")

else:
    print(res)
```

```python
age = int(input("please enter your age"))

if age < 0 or age > 100:
    raise "Age Error"
```

```python
age = int(input("please enter your age"))
try:
    if age < 0 or age > 100:
        raise "Age Error"
except Exception as e:
    print(e)
```

```python
class AgeException(Exception):
    pass
```

```python
age = int(input("please enter your age"))
try:
    if age < 0 or age > 100:
        raise AgeException("Age can not be less than 0 or greater than 100")
except Exception as e:
    print(e)
```
