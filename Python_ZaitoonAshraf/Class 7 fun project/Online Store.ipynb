```python
store = [
{'item':'rice', 'price':300, 'quantity':1000},
{'item':'floor', 'price':200, 'quantity':1000},
{'item':'soap', 'price':200, 'quantity':1000},
{'item':'shampo', 'price':1200, 'quantity':1000},
{'item':'surf', 'price':300, 'quantity':1000},
{'item':'milk', 'price':400, 'quantity':1000},
{'item':'tea', 'price':480, 'quantity':1000},
{'item':'egg', 'price':1000, 'quantity':1000},
{'item':'sugar', 'price':5000000, 'quantity':1000},
{'item':'bread', 'price':190, 'quantity':1000}
]


cart=[]
product = input('Kia chahiye hai apko?')
for p in store:
    if p['item'] == product:
        print("Mill jaega")
        print(f'''{product} {p['price']}''')
        quant = int(input("kitni chahiye"))
        if quant <=p['quantity']:
            amount = p['price'] * quant
            p['quant']-=quant
            cart.append({"item":product, "amount":amount, 'quantity':quant})
            break
        else:
            print(f''' {product}{quant} availble nahi ha ''')
        
        
        # break
else:
    print("not availble")
    
```
