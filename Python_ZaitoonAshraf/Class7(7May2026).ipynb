```python
available_toppings = ['Mushroom', 'Olives', 'Onion', 'Pepricca', 'Cheese', 'Capcicam']
applied_toppings=[]
while True:
    topping = input("Please enter your topping or q to exit")
    if topping!='q':
        if topping in available_toppings:
            applied_toppings.append(topping)
        else:
            print("Sorry! Requested topping not in list")
    else:
        print("Thank you for ordering pizza")
        break
applied_toppings        
```

```python
store =[{'item':'rice', 'price':300, 'quantity':1000},
       {'item':'floor', 'price':200, 'quantity':1000},
       {'item':'oil', 'price':600, 'quantity':1000},
       {'item':'soap', 'price':200, 'quantity':1000},
       {'item':'shampo', 'price':1200, 'quantity':1000},
       {'item':'sugar', 'price':170, 'quantity':1000},
       {'item':'surf', 'price':300, 'quantity':1000},
       {'item':'milk', 'price':240, 'quantity':1000},
       {'item':'tea', 'price':700, 'quantity':1000},
       {'item':'egg', 'price':25, 'quantity':1000},
       {'item':'bread', 'price':100, 'quantity':1000}]
```

```python
cart = []
product = input("Kia chahye?   ")
for p in store:
    if p['item']==product:
        print('Mil jaega')
        print(f"{product} {p['price']} rupey ki hy") 
        quant = int(input("Kitni chahye"))
        if quant<=p['quantity']:
            amount = p['price']* quant
            p['quantity']-=quant
            cart.append({"item":product, "amount":amount, 'quantity':quant})
            break            
        else:
            print(f"{product}{quant} avaialble nahi hy")
            break
else:
    print("Not available")

cart
```

```python
cart = []
total= 0
while True:
    product = input("Kia chahye? or press q to exit  ")
    if product!='q':
        for p in store:
            if p['item']==product:
                print('Mil jaega')
                print(f"{product} {p['price']} rupey ki hy") 
                quant = int(input("Kitni chahye"))
                if quant<=p['quantity']:
                    amount = p['price']* quant
                    p['quantity']-=quant
                    cart.append({"item":product, "amount":amount, 'quantity':quant})
                    break                    
                else:
                    print(f"{product}{quant} avaialble nahi hy")
                    break
        else:
            print("Not available")
    else:
        print("Thank you for shopping with us")
        break

for c in cart:
    total+=c['amount']
    
print(f"""
            #########Invoice###########

       Items: {[i['item'] for i in cart]}
       Prices: {[i['amount'] for i in cart]}
       Total : {total}

""")
```
