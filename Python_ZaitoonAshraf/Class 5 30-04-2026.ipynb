# Nested Conditions
    We have to buy somethings. (Laptops)
     - Hp
     - Corei7
     - SSD 1Tb
     - Ram 16gb
     - Generation 11th
     

```python


brand = input("Hp Laptop Are Availble ?").lower()
if brand == 'hp':
    process = input("Corei7 availble hai?").lower()
    if process == 'i7':
        ssd = input("kia apke pass 1tb ssd hai?").lower()
        if ssd == '1tb':
            ram = input("Kia apke pass 16gb ram hai?").lower()
            if ram == '16':
                Generation = input("apke pass 11th process hai").lower()
                if Generation == 'g':
                    print("theek hai yehi wala system dedo")
                else:
                    print("mujhay to 11th generation chahiye")
            else:
                print("mujay to 16gb ram Chahiye")     
        else:
            print("mujhay 1tb ssd chahiye")
    else:
        print("mujhay core i 7 chaiye")
else:
    print("mujahy to HP he chahiye Thanks!")
```
