Loops Solutions


```python
numbers = [1, -2 ,3 ,-4,6,7,-7,-8,9,-10]
```

```python
positive_number_count =0
for num in numbers:
    if num >0:
        positive_number_count += 1
print("final count of positive number is: ", positive_number_count)
```

```python
n = 100
sum_even = 0

for i in range(1, n+1):
    if i%2 == 0:
       sum_even +=1

print(sum_even)
```

solution 5 calculate factorial of a number using white loop


```python
number = 5
factorial = 1

while True:
    factorial *= number
    number <=1
print(factorial)
```
